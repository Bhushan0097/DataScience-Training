#### Q10.Consider this class data as a sample for population of Nagpur , estimate the true mean of the height of population with confidence interval for confidence level 90% and 92%.

```python
from scipy.stats import norm
import math

mu = classdata['ht_cm'].mean()
sigma = classdata['ht_cm'].std()
n = len(classdata['ht_cm'])

#for 90% 
z90 = norm.ppf(0.05)
E90 = z90*(sigma/math.sqrt(n)) 
CI90l = mu - E90
CI90u = mu + E90

print('For Confidence level 90%% \nConfidence Interval :\n \22 Lowerlimit ---> %2.4f'%CI90l ,'\n \22  Upperlimit ---> %2.4f'%CI90u)

#for 92% 
z92 = norm.ppf(0.04)
E92 = z92*(sigma/math.sqrt(n)) 
CI92l = mu - E92
CI92u = mu + E92


print('\nFor Confidence level 92%% \nConfidence Interval :\n \22 Lowerlimit ---> %2.4f'%CI92l ,'\n \22  Upperlimit ---> %2.4f'%CI92u)
```

#### Q11. Consider this data as a sample for survey that what type marriage the youngster prefer now a days. Estimate what is the proportion of population prefer arrange marriages with interval for confidence level of 85% and 90%.

```python
Yes = classdata.groupby('ar_ma')
ar_ma = Yes.get_group('Yes')
```

```python
ar_ma
```

confidence Interval - Sample interval 
 
 Ps +/- Z{sqrt[Ps(1-Ps)/n]}

```python
p_s = (ar_ma['ar_ma'].count())/100
n = len(classdata)
ps = ar_ma['ar_ma'].count()/n
print('ps =',p_s, 'and n =',n)
print('Proportion of population ',(ar_ma['ar_ma'].count()*100/n))

data = (ps*(1-ps))/n

#for 85%
z85 = norm.ppf(0.075)
CI85l = ps + z85*(math.sqrt(data))
CI85u = ps - z85*(math.sqrt(data))
print('Confidence interval for 85%%  is  %2.4f'%CI85l,' to %2.4f'%CI85u)

#for 90%
z90 = norm.ppf(0.05)
CI90l = ps  + z90*(math.sqrt(data))
CI90u = ps - z90*(math.sqrt(data))
print('Confidence interval for 90%%  is  %2.4f'%CI90l,' to %2.4f'%CI90u)
```

#### Q12. Consider this data as a sample for survey that which hospital, government or private the youngster prefer now a days for medical treatment. Estimate what is theproportion of population prefer government hospital with interval for confidence level of88% and 94%.

```python
Yes = classdata.groupby('govt_hos')
govt_hos = Yes.get_group('Yes')
```

```python
govt_hos
```

```python
p_s = (govt_hos['govt_hos'].count())/100
n = len(classdata)
ps = govt_hos['govt_hos'].count()/n
print('ps =',p_s, 'and n =',n)
print('Proportion of population %2.4f' %(govt_hos['govt_hos'].count()*100/n))

data = ps*(1-ps)/n

#for 88%
z88 = norm.ppf(0.06)

CI88l = ps + z88*(math.sqrt(data))
CI88u = ps - z88*(math.sqrt(data))
print('Confidence interval for 85%%  is  %2.4f'%CI88l,' to %2.4f'%CI88u)

#for 94%
z94 = norm.ppf(0.03)
CI94l = ps + z94*(math.sqrt(data))
CI94u = ps - z94*(math.sqrt(data))
print('Confidence interval for 90%%  is  %2.4f'%CI94l,' to %2.4f'%CI94u)
```
