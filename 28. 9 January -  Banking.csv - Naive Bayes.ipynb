```python
import pandas as pd
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB

bank = pd.read_csv('bank.csv')
bank
```

```python
bank.describe()
```

```python
X = pd.DataFrame(bank[['age','balance','day','duration','campaign','pdays','previous']])
y = pd.DataFrame(bank['deposit'])
```

```python
from sklearn import linear_model
from sklearn import datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20,random_state = 0)
```

```python
#Creating our Naive Bayes Model using Sklearn
from sklearn.metrics import classification_report,confusion_matrix
model = GaussianNB()
model.fit(X_train,y_train)
#making prediction 
expected = y_test
predicted  = model.predict(X_test)
print(metrics.classification_report(expected,predicted))
#print(metrics.confusion_matrix((expected,predicted)))

from sklearn.metrics import accuracy_score
accur = accuracy_score(expected,predicted)
print('Test Accuracy :{:.6f}'.format(accur))
```
