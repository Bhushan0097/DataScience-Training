    Population Propotion
    • H0 : P less than eqaul to 0.10

```python
from scipy.stats import norm
pbar = 13/100 #sample propotion
pzero = 0.10 #hypothesised value
n =100
mu,sigma = 0,1

z1 = pbar - pzero
z2 = (pzero*(1-pzero)/n)**0.5

z13 = z1/z2
#The are on the left side of STD, NOR

p13 = norm.cdf(z13,mu,sigma)
#since it is upper tale test
print('Upper tale P value : %2.4f' %(1-p13))
```

    p-value 0.1587 is greater than 0.05. We accept H0 and hence should not go for National

```python
from scipy.stats import norm
import pandas as pd

def propotion(pro):
    pbar = pro/100
    pzero = 0.10 #hypothesised value
    n =100
    mu,sigma = 0,1
    
    z1 = pbar - pzero
    z2 = (pzero*(1-pzero)/n)**0.5

    z = z1/z2
    
    #The are on the left side of STD, NOR
    p = norm.cdf(z,mu,sigma)

    data = dict()
    data['pbar'] = pro
    data['z'] = z
    data['Value of p']= 1-p
    
    if (1 -p ) < 0.05:
        data['Accept'] = 'H1'
    else :
        data['Accept'] = 'H0'
    
    return data 


df100 = pd.DataFrame({'pbar':[],'z':[],'Value of p': [],'Accept': []})
for pro in range(13,20):
    df100= df100.append(propotion(pro),ignore_index = True)
    
    
```

```python
from scipy.stats import norm
import pandas as pd

def propotion(pro):
    pbar = pro/500
    pzero = 0.51 #hypothesised value
    n =500
    mu,sigma = 0,1
    
    z1 = pbar - pzero
    z2 = (pzero*(1-pzero)/n)**0.5

    z = z1/z2
    
    #The are on the left side of STD, NOR
    p = norm.cdf(z,mu,sigma)

    data = dict()
    data['pbar'] = pro
    data['z'] = z
    data['Value of p']= 1-p
    
    if (1 -p ) < 0.01:
        data['Accept'] = 'H1'
    else :
        data['Accept'] = 'H0'
    
    return data 


df100 = pd.DataFrame({'pbar':[],'z':[],'Value of p': [],'Accept': []})
for pro in range(232,501):
    df100= df100.append(propotion(pro),ignore_index = True)
    
    
```
