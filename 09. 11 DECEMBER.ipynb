```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cars = pd.read_csv('E:\\python\\Assingments\\Cars_data.csv') 
cars.head()
```

```python
cars_gr = cars.groupby('company')
mileage = cars_gr['company','average-mileage'].mean()
print(mileage)
```

```python
cols = ['violet','indigo','blue','green','yellow','orange','red','grey','gold','silver','brown','black','pink','maroon','darkred','lightblue']

plt.title('Bar Chart for car mileage')
plt.ylabel('Mileage in km per liter')
plt.bar(mileage.index,mileage['average-mileage'],align ='center',color = cols)
plt.xticks(rotation = 90)

plt.show()
```

```python
#count of number of cars companywise
car_count = cars['company'].value_counts()
#print(car_count)

plt.pie(car_count,startangle = 90, autopct = '%2.2f%%')
plt.title('Car Count')

#plt.legend(car_count.index, loc =3)

plt.legend(car_count.index, loc ='upper right' ,ncol = 1)
plt.show()
```
