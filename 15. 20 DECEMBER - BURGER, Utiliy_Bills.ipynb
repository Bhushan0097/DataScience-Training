• An Insurance company wants to determine how its annual operating costs depend on the number of home insurance and automobile insurance policies that have been written. The file Insure_auto.csv contains relevant data from 10 branches of insurance company

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
insure = pd.read_csv('E://python//Insure_auto.csv')
#X = pd.DataFrame(insure['Home'])
print(insure.info)

X = insure[['Home','Automobile']]
y = insure[['Operating_Cost']]

lm = linear_model.LinearRegression()
model = lm.fit(X,y)
print('coeficent',lm.coef_)
print('Intercept',lm.intercept_)
print('R square',lm.score(X,y))

test_data =  pd.read_csv('E://python//Insure_auto2.csv')
df_test = pd.DataFrame(test_data, columns = ['Home','Automobile'],)
predict = lm.predict(df_test)
print(predict)
test_data['Operating_Cost'] = predict
```
