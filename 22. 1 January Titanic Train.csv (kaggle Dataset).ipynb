<font face="verdana">5. Data Binning : Create a extra column <b> â€˜Age Groupâ€™</b> by using binning concept. Use the following table for creating category.<br>
  â€¢  Child - 0 to 12 <br>
  â€¢  Adolescence 13 to 18 <br>
  â€¢  Adult - 19 to 40 <br>
  â€¢  Middle Age 41 to 60 <br>
  â€¢  Senior 61 to 75 <br>
  â€¢  Very Senior 76 and above <br>

```python
age2 = pd.qcut((train['Age']), q=[0,0.0764,0.1561,0.831459,0.9752,0.9991,1] ,labels =['Child','Adolescence','Adult','Middle Age','Senior','Very Senior'],duplicates = 'drop')
```

```python
age2
```

```python
age100 = pd.DataFrame({'age': train['Age'], 'Age_class':age2})
```

<font face="verdana">6. Data Visualization : <br>
a. Passenger survived vs Gender of passengers

```python
gender = train.groupby('Sex')

#FEMALE
gender_female = gender.get_group('female')
print('No of Female :',gender_female['Sex'].count())
print('No of Female Survived :',gender_female['Survived'].sum())
xx = (gender_female['Sex'].count() - gender_female['Survived'].sum())
print('No of Female NOT Survived :',xx)


#MALE
gender_male = gender.get_group('male')
print('\nNo of male:',gender_male['Sex'].count())
print('No of male Survived :',gender_male['Survived'].sum())
xy =(gender_male['Sex'].count() - gender_male['Survived'].sum())
print('No of male NOT Survived :',xy)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Gender',fontsize = 15)
ax.set_ylabel('Survived',fontsize = 15)
ax.set_title('Gender v/s Survived',fontsize = 20)

bars = ['Female','Male']
height = (gender_female['Survived'].sum(), gender_male['Survived'].sum())
color = ['#DC7633','#85929E']

plt.ylim(0,400)
y_pos = np.arange(len(bars))
plt.xticks(y_pos, bars)

for y_pos, height,color in zip(y_pos, height,color):
    plt.bar(y_pos, height, color =color)

plt.legend(bars)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8,8))
train.boxplot(column ='Age')
```

```python
import numpy as np
Q1 = np.quantile(train['Age'],0.25)
Q2 = np.quantile(train['Age'],0.50)
Q3 = np.quantile(train['Age'],0.75)

print('Q1 : ',Q1,'\nQ2 : %2.4f  '%Q2,'\nQ3 : ',Q3,)
```

<font face="verdana"> Data Outliers: <br>
a. Calculate the limits of mild & extreme outliers for the age

```python
IQR = Q3 - Q1

lower =  Q1- 1.5*IQR
upper = Q3 + 1.5*IQR
lof = Q1 - 3 *IQR
uof = Q3 + 3 *IQR

print('Lower inner fence :' ,lower)
print('Upper inner fencer :' ,upper)
print('Lower outer fencer :' ,lof)
print('upper outer fencer :' ,uof)

print("\nMild Outlier  : ", lof)
print("Extreme Outlier : ", uof)
```

<font face="verdana"> b. Check if any data points as outliers. <br>
c. If not create a point as out layer and check if shown in box plot.

```python
outlier = train[(train['Age']< lower )| (train['Age']>upper)]
print(outlier['Age'])        
```

```python
outlier.describe()
```

<font face="verdana"> 8.Data Out layers: <br>
a. Calculate lower and upper limit of out layers data points as THREE sigma imits. <br>

```python
outlier.shape
```

```python
x =  outlier['Age'].mean() - 3*outlier['Age'].std()
y  = outlier['Age'].mean() + 3*outlier['Age'].std()
```

```python
print(" Lower Limit :" ,x)
print(" Upper Limit :",y)
```

```python
#Extreme OUTLIER

ex_out = train[(train['Age']< lof )| (train['Age']>uof )]
print(ex_out['Age'])   
```

<font face="verdana">b. Check if points fall outside this limit and drop them.<br>

```python
train.drop(train[(train['Age']< x )| (train['Age']>y)].index, inplace = True)
```
