```python
import numpy as np
import pandas as pd

owners = pd.DataFrame({'Names':['John','Johny','Janardhan','Amar','Akbar','Anathony'], 'Cars': ['Maruti','Honda','Honda','Toyota','Maruti','Toyota']})
print(owners)
```

```python
print(owners.info())
```

```python
print(owners.describe())
```

```python
#Example of encoding
from sklearn import preprocessing

#create the lalbelencoder object
le = preprocessing.LabelEncoder()

#convert the categorical columns into numeric
encoded_values = le.fit_transform(owners.Cars)
print(encoded_values)

owners['Code'] = le.fit_transform(owners.Cars)
print(owners)
```

```python
print(owners.info())
```

```python
#creating dummies
new_owners = pd.get_dummies(owners.Cars)
print(new_owners)
new_owners = pd.concat([owners,new_owners],axis =1)
print(new_owners)
```

```python
print(new_owners.info())
```
