```python
#PCA Implementation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import DateSet Iris

df = pd.read_csv("E:\\python\\iris.csv")
df
```

```python
df['class'].unique()
```

```python
#Split the data into feartures and target

X = df.iloc[:,:-1]
y = df.iloc[:,4].values
```

```python
print(X)
print(y)
```

```python
from sklearn import linear_model
from sklearn import datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20,random_state = 0)
```

```python
##Using Min Max Scaler
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

```python
#make prediction for test data
y_pred = classifier.predict(X_test)
```

```python
#Evaluating the Algorithm
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Evaluation

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
accur = accuracy_score(y_test, y_pred)
print('Test Accuracy :{:.6f}'.format(accur))
```

### Using StandardScaler

```python
#Features scaling mean = 0 & varianace = 1

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

X1 = scaler.transform(X)
print(X1)
```

```python
#PCA projection to 2D
#The Original data has four columns and the new data has 2
#You cannot attach any meaning to these two newly created columns

from sklearn.decomposition import PCA

pca = PCA(n_components=4)
pca2 = pca.fit_transform(X1)
pca2
```

```python
principalDF = pd.DataFrame(data = pca2, columns = ['PC1','PC2','PC3','PC4'])
principalDF
```

```python
finalDf = pd.concat([principalDF,df[['class']]],axis =1)
finalDf.head()
```

```python
X = finalDf[['PC1','PC2','PC3','PC4']]
y = pd.DataFrame(finalDf['class'])
print(X.head())
print(y.head())
```

```python
from sklearn import linear_model
from sklearn import datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

```python
#make prediction for test data
y_pred = classifier.predict(X_test)
```

```python
#Evaluation

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
accur = accuracy_score(y_test, y_pred)
print('Test Accuracy :{:.6f}'.format(accur))
```

```python
#we randomly chose 5 as the K value and it just happen to result in 100%
#One way to help you find  the best value of K is to plot the graph of k
#and the corresponding error rate for the dataset.
#In this section, we will plot the mean error for the predicted values of
# set for all the k values between 1 and 40
from sklearn.neighbors import KNeighborsClassifier

#calculating error for K values between 1 and 40
error = []

for i in range(1,len(y_test)):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train.values.ravel())
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i[i] != y_test))
```

```python
import matplotlib.pyplot as plt
#The next step is to plot the error values against K values.

plt.figure(figsize=(12,6))
plt.plot(range(1,len(y_test)),error,color = 'red',linestyle= 'dashed',marker = 'o',markerfacecolor= 'blue',markersize = 10 )
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
from sklearn.model_selection import cross_val_score
error = []
for i in range(1,90):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn,X_train, y_train.values.ravel(), cv = 10)
    error.append(1-score.mean())
```

```python
#The next Step is to plot the error values against K values.
import matplotlib.pyplot as plt
#The next step is to plot the error values against K values.

plt.figure(figsize=(12,6))
plt.plot(range(1,90),error,color = 'red',linestyle= 'dashed',marker = 'o',markerfacecolor= 'blue',markersize = 10 )
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```
