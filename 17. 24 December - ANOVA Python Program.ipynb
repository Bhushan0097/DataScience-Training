```python
#Null hypothesis (H0) the mean of sample 1 is same as mean of sample 2

from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv('E://Python//Anova_py.csv')
print(df1)

st,p = stats.ttest_rel(df1['sample 1'],df1['sample 2'])
print('\nStatistic value =%6.4f and pvalue = %8.5f'%(st,p))
#Since p value is less than 0.05, we reject Null Hypothesis

st,p = stats.ttest_rel(df1['sample 2'],df1['sample 3'])
print('\nStatistic value =%6.4f and pvalue = %8.5f'%(st,p))
#Since p value is more than 0.05, we reject Null Hypothesis
#Both Sample have same mean ,both sample are similar

st,p = stats.ttest_rel(df1['sample 1'],df1['sample 3'])
print('\nStatistic value =%6.4f and pvalue = %8.5f'%(st,p))
#Since p value is less than 0.05, we reject Null Hypothesis
#Both Sample have same mean ,both sample are similar
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv('E://Python//mtcars1.csv')
data
```

```python
import pandas
import seaborn as sns
import numpy as np
import statsmodels.api as sm 
from  statsmodels.formula.api import ols
import statsmodels.stats.multicomp

formula = 'mpg ~ hp +am'   # no interaction
model = ols(formula,data).fit()
aov_table = statsmodels.stats.anova.anova_lm(model,typ =1 )
print(aov_table)
```
